```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

datos = sns.load_dataset('penguins')
datos.head()
```

```python
print(datos['species'].unique(), datos['island'].unique(), "\n\n", datos['bill_length_mm'].unique(), "\n\n" ,datos['bill_depth_mm'].unique(), "\n\n", datos['flipper_length_mm'].unique(), "\n\n", datos['body_mass_g'].unique(), "\n\n", datos['sex'].unique())
```

```python
"""
PREGUNTA 7:
La columna 'sex' se refiere al sexo de cada pingüino (macho / hembra).
La columna 'body_mass_g' se refiere a la masa de cada pingüino, en g.
"""
```

```python
datos = datos.dropna()

male = datos[datos.sex.str.match(r'^Male$')]
male.head()

female = datos[datos.sex.str.match(r'^Female$')]
female.head()

m = male['body_mass_g']
f = female['body_mass_g']
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, density=False, bins=40)

plt.hist(m, label="Machos", **kwargs)
plt.hist(f, label="Hembras",**kwargs)
plt.legend();
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, density=False, bins=40)

plt.hist(m, label="Machos", **kwargs)
plt.hist(f, label="Hembras",**kwargs)
plt.legend()
plt.annotate('Se obseva que los\n machos pesan más',
            xy=(1, 0), xycoords='axes fraction',
            xytext=(-10, 140), textcoords='offset pixels',
            horizontalalignment='right',
            verticalalignment='bottom');
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, density=False, bins=40)

plt.hist(m, label="Machos", **kwargs)
plt.hist(f, label="Hembras",**kwargs)
plt.legend()
plt.annotate('Las hembras \npesan menos',
            xy=(0.1, 0.4), xycoords='axes fraction',
            xytext=(40, 100), textcoords='offset pixels',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='right',
            verticalalignment='bottom');
```

```python
sns.heatmap(datos.corr())
```

```python
sns.pairplot(datos, hue='sex')
```
